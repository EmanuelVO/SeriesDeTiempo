```python
import yfinance as yf
import pandas as pd
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

```python
ibm_wmt = yf.download(tickers='ibm wmt', period='5y', interval='1mo', rounding=True)
ibm_wmt
```

```python
ibm_wmt = ibm_wmt['Close']
ibm_wmt
```

```python
ibm_wmt['IBM'].plot()
ibm_wmt['WMT'].plot()
plt.legend(['IBM', 'WALMART'])
plt.title('Comparacion de precios IBM-Walmart')
plt.show()
```

```python
rend_ibm_wmt = ibm_wmt.pct_change()
rend_ibm_wmt = rend_ibm_wmt.dropna()
correlacion = rend_ibm_wmt['IBM'].corr(rend_ibm_wmt['WMT'])
print(correlacion)
```

```python
plt.scatter(rend_ibm_wmt['IBM'], rend_ibm_wmt['WMT'])
plt.show()
```

```python
# Prueba de caminata aleatoria
# H0: se tiene un proceso de caminata aleatoria
# Ha: no se teiene un proceso de caminata aleatoria

from statsmodels.tsa.stattools import adfuller
import numpy as np
```

```python
# Prueba Dickey-Fuller para los precios de IBM
ibm = ibm_wmt['IBM']
resultado = adfuller(ibm)
print('El valor p de la prueba es: ' + str(resultado[1]))
```

```python
# Prueba Dickey-Fuller para los precios de WALMART
wmt = ibm_wmt['WMT']
resultado = adfuller(wmt)
print('El valor p de la prueba es: ' + str(resultado[1]))
```

```python
# Generacion de la funcion de autocorrelacion
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
```

```python
acf_array = acf(ibm)
print(acf_array)
```

```python
plot_acf(ibm, alpha=0.05)
plt.title('IBM')
plt.show()
```

```python
acf_array = acf(wmt)
print(acf_array)
```

```python
plot_acf(wmt, alpha=0.05)
plt.title('WALMART')
plt.show()
```

```python
# Graficacion de la media movil para IBM
rolling_ibm = ibm.rolling(30).mean()
ibm.plot()
rolling_ibm.plot()
plt.title('Precios historicos de IBM vs Precios moviles a 30 dias')
plt.legend(['IBM', 'Media movil'])
plt.show()
```

```python
# Descomposicion de la serie de tiempo
ibm.plot(figsize=(16,8)).grid()
```

```python
# Graficacion de la media movil para WALMART
rolling_wmt = wmt.rolling(30).mean()
wmt.plot()
rolling_wmt.plot()
plt.title('Precios historicos de WALMART vs Precios moviles a 30 dias')
plt.legend(['WALMART', 'Media movil'])
plt.show()
```

```python
# Descomposicion de la serie de tiempo
wmt.plot(figsize=(16,8)).grid()
```

```python
import statsmodels.api as sm
decomposed_ibm = sm.tsa.seasonal_decompose(ibm, period=30)
decomposed_wmt = sm.tsa.seasonal_decompose(wmt, period=30)
```

```python
figure = decomposed_ibm.plot()
plt.grid()
plt.show()
```

```python
figure = decomposed_wmt.plot()
plt.grid()
plt.show()
```

# IBM

```python
# Prediccion simple mediante promedios moviles
ibm.shape
```

```python
len_train = int(60 * 0.8)
len_train
```

```python
# Definicion de grupos de entremaniento y de prueba
train = ibm[0:len_train]
test = ibm[len_train:]
```

```python
len(train)
```

```python
len(test)
```

```python
y_pred = ibm.copy()
y_pred = pd.DataFrame(y_pred)
```

```python
y_pred['pronostico'] = ibm.rolling(30).mean()
y_pred[0:35]
```

```python
y_pred_m = y_pred.copy()
y_pred_m['pronostico'][29] = 'NaN'
for contador in range(30, 60):
    y_pred_m['pronostico'][contador] = y_pred['pronostico'][contador-1]
y_pred_m[0:35]
```

```python
y_pred_m = pd.DataFrame(y_pred_m['pronostico'][48:])
y_pred_m
```

```python
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(y_pred_m['pronostico'], label='Prediction', color='red')
plt.legend(loc='best')
plt.title('Pronosticos mediante medias moviles')
plt.show()
```

```python
test = pd.DataFrame(test)
test
```

```python
frames = [test, y_pred_m]
result = pd.concat(frames, axis=1, join='inner')
result.columns = ['Observado', 'Pronosticado']
result
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(test, y_pred_m)).round(2)
mape = np.round(np.mean(np.abs((test['IBM'] - y_pred_m['pronostico']) / test['IBM'])) * 100, 2)
```

```python
print('RMSE = ', rmse, 'MAPE = ', mape)
```

```python
# Pronostico para el siguiente 
y_pred.tail(1)['pronostico']
```
